```python
missing_values = train.isnull().sum()
missing_values[missing_values > 0]
```

```python
less = missing_values[missing_values < 1000].index
over = missing_values[missing_values >= 1000].index
over
```

```python
numeric_features = train[less].select_dtypes(include=['number']).columns
train[numeric_features] = train[numeric_features].fillna(train[numeric_features].median())
```

```python
train[less].select_dtypes(include=['number']).isnull().sum()
```

```python
kategorical_features = train[less].select_dtypes('object').columns
kategorical_features
```

```python
train[kategorical_features] = train[kategorical_features].fillna(train[kategorical_features].mode().iloc[0])
```

```python
train[kategorical_features].isnull().sum()
```

```python
#menghapus kolom dengan nterlalu banyak nilai yang hilang
df = train.drop(columns=over)
df
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(15 , 30))
for i, feature in enumerate(numeric_features):
  plt.subplot(10, int(round(len(numeric_features) / 8)), i + 1)
  sns.boxplot(y=df[feature], orient = 'v')
  plt.tight_layout()
  plt.title(feature)
```

```python
q1 = df[numeric_features].quantile(0.25)
q3 = df[numeric_features].quantile(0.75)

iqr = q3 - q1
```

```python
condition = ~((df[numeric_features] < (q1 - 1.5 * iqr) | (df[numeric_features] > (q3 + 1.5 * iqr))).any(axis=1))
df_filtered_numeric = df.loc[condition, numeric_features]
df_filtered_numeric
```

```python
categorical_features = df.select_dtypes(include=['object']).columns
df = pd.concat([df_filtered_numeric, df.loc[condition, categorical_features]], axis = 1)
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 15))
for i, feature in enumerate(df_train.columns):
    plt.subplot(7, 5, i + 1)
    sns.boxplot(x=df_train[feature], palette = 'viridis')
plt.tight_layout()
plt.show()
```

```python
q1 = df_train.quantile(0.25)
q3 = df_train.quantile(0.75)

iqr = q3 - q1
lower = q1 - 1.5 * iqr
upper = q3 + 1.5 * iqr 

condition = ~((df_train < lower) | (df_train > upper)).any(axis=1)
df = df_train.loc[condition, df_train.columns]
df
```

```python
from sklearn.preprocessing import StandardScaler

numeric_features = df.select_dtypes('number').columns
numeric_features

scaler = StandardScaler()
df[numeric_features] = scaler.fit_transform(df[numeric_features])

df
```

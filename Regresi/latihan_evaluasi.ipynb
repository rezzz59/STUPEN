```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# 1. DATA DARI MODUL
# Harga Asli (y_true)
y_asli = np.array([100, 150, 200, 250, 300])
# Harga Prediksi (y_pred) - Hasil hitungan manual y = 50 + 2.5x
y_prediksi = np.array([175, 200, 225, 250, 275])

# 2. HITUNG PAKAI SCIKIT-LEARN
mae = mean_absolute_error(y_asli, y_prediksi)
mse = mean_squared_error(y_asli, y_prediksi)
rmse = np.sqrt(mse) # Scikit-learn gak punya fungsi langsung RMSE, jadi kita akar manual
r2 = r2_score(y_asli, y_prediksi)

# 3. TAMPILKAN RAPOR
print("=== RAPOR MODEL REGRESI ===")
print(f"MAE  (Rata-rata Meleset) : {mae} (Harus 35)")
print(f"MSE  (Kuadrat Error)     : {mse} (Harus 1875)")
print(f"RMSE (Akar MSE)          : {rmse:.1f} (Harus 43.3)")
print(f"R2 Score (Kecerdasan)    : {r2} (Harus 0.625)")

# Interpretasi Singkat
print("\n=== INTERPRETASI ===")
print(f"Rata-rata tebakan harga rumah meleset sebesar Rp {mae} Juta.")
print(f"Skor kecerdasan model adalah {r2*100}%")
```

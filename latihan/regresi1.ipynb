```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. TARIK DATA DARI SUMBER PUBLIK (UCI Repository)
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"

# Karena file aslinya polosan (gak ada header), kita buat daftar nama kolomnya dulu
cols = [
    "symboling", "normalized-losses", "make", "fuel-type", "aspiration",
    "num-of-doors", "body-style", "drive-wheels", "engine-location", "wheel-base",
    "length", "width", "height", "curb-weight", "engine-type",
    "num-of-cylinders", "engine-size", "fuel-system", "bore", "stroke",
    "compression-ratio", "horsepower", "peak-rpm", "city-mpg", "highway-mpg", "price"
]

# Baca data csv
df_auto = pd.read_csv(url, names=cols)

# 2. CEK KEKOTORAN DATA
print("--- Info Dataset ---")
print(df_auto.info())

print("\n--- Intip 10 Baris Pertama ---")
display(df_auto.head(10))

print("\n--- Cek Statistik Dasar ---")
# Hati-hati, describe() cuma akan menghitung kolom yang dianggap ANGKA oleh Pandas
display(df_auto.describe())
```

```python
import numpy as np

df_auto.replace('?', np.nan, inplace=True)
df_auto
```

```python
df_auto.isnull().sum()
```

```python
df_auto.info()
```

```python
numeric_cols = ['normalized-losses', 'bore', 'stroke', 'horsepower', 'peak-rpm', 'price']


for col in numeric_cols:
    df_auto[col] = pd.to_numeric(df_auto[col], errors='coerce')
df_auto.info()
```

```python
df_auto.isnull().sum()
```

```python
df_auto = df_auto.dropna(subset=['price'], axis=0)
```

```python
df_auto.isnull().sum()
```

```python
df_auto.select_dtypes(include=['float64', 'int64']).mean()
```

```python
df_auto.info()
```

```python
numerical = df_auto.select_dtypes(include='number').columns
numerical
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 

plt.figure(figsize=(13, 30))
for i, col in enumerate(numerical):
  plt.subplot(6, 4, i + 1)
  sns.histplot(df_auto[col])
plt.tight_layout()
plt.show()
```

```python
df_auto['symboling'] = df_auto['symboling'].fillna(df_auto['symboling'].mean())
df_auto['bore'] = df_auto['bore'].fillna(df_auto['bore'].mean())
df_auto['stroke'] = df_auto['stroke'].fillna(df_auto['stroke'].mean())
```

```python
df_auto['normalized-losses'] = df_auto['normalized-losses'].fillna(df_auto['normalized-losses'].median())
```

```python
df_auto['num-of-doors'] = df_auto['num-of-doors'].fillna(df_auto['num-of-doors'].mode()[0])
```

```python
df_auto.info()
```

```python
korelasi = df_auto.select_dtypes('number').corr()

plt.figure(figsize=(12, 8))
sns.heatmap(korelasi, annot=True, cmap='coolwarm')
plt.title('Peta korelasi fitur pada mobil vs harga')
```

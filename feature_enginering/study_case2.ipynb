```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_classification
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder, StandardScaler
from imblearn.over_sampling import SMOTE
from collections import Counter
```

```python
#generate dataset 
X, y = make_classification(n_samples=1000, n_features=15, n_informative=10, n_redundant=2, n_clusters_per_class=1, weights=[0.9], flip_y=0, random_state=42)
 
df = pd.DataFrame(X, columns=[f'fitur_{i}' for i in range(1, 16)])
df['target'] = y
df['fitur_12'] = np.random.choice(['A', 'B', 'C'], size=1000)
df['fitur_13'] = np.random.choice(['X', 'Y', 'Z'], size=1000)

df
```

```python
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
numerical = x.drop(['fitur_12', 'fitur_13'], axis=1)
rf_model.fit(numerical, y)

#mendapatkan fitur penting
importance = rf_model.feature_importances_
indices = np.argsort(importance)[::-1]

treshold = 0.05
important_feature_indices = [i for i in range(len(importance)) if importance[i] >= treshold]
important_feature_indices

print('Fitur yang didapatkan dari ambang batas treshold')
for i in important_feature_indices:
    print(f'{numerical.columns[i]} : {importance[i]}')

important_features = numerical.columns[important_feature_indices]
important_features = numerical[important_features]
print(important_features)
print("Dimensi data pelatihan dengan fitur penting :", important_features.shape)
```

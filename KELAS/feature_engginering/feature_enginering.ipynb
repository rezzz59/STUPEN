```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.feature_selection import RFE 
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_wine
```

```python
data = load_wine()
x, y = data.data, data.target
```

```python
df = pd.DataFrame(x, columns=data.feature_names)
df['Target'] = y
df
```

```python
df.describe()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
x_train
```

# FILTER METHODS

```python
#menggunakan selectKbest 
filter_selector = SelectKBest(score_func= chi2, k=2) #memilih 2 fitur terbaik
x_train_filter = filter_selector.fit_transform(x_train, y_train)
x_test_filter = filter_selector.transform(x_test)

print(f"Jumlah fitur setelah seleksi {x_train_filter.shape[1]}")

#mendapatkan indeks fitur yang terpilih
selected_indeces = filter_selector.get_support(indices=True)
selected_indeces
```

```python
#melihat nama fitur  yang terpilih
selected_features = [data.feature_names[i] for i in selected_indeces]
selected_features
```

```python
#melihat skor masing-masing fitur
```

# WRAPPER METHOD

```python
#menggunakan RFE(Recursive  Feature Selection )
model = LogisticRegression(solver='lbfgs', max_iter= 5000)
rfe_selector = RFE(model, n_features_to_select=2)
x_train_rfe = rfe_selector.fit_transform(x_train, y_train)
x_test_rfe = rfe_selector.transform(x_test)

result = rfe_selector.get_support(indices=True)
result
```

# Embedded Methods

```python
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(x_train, y_train)

importance = rf_model.feature_importances_
importance
```

```python
indices = np.argsort(importance)[::-1]
indices
```

```python
treshold = 0.05
important_feature_indices = [i for i in range(len(importance)) if importance[i] >= treshold]
important_feature_indices
```

```python
x_important = x_train[:, important_feature_indices]
x_test_important = x_test[:, important_feature_indices]

```

```python
#mengecek fitur yang dipilih 
for colum, i in enumerate(important_feature_indices):
    print(f'{data.feature_names[i]} : {importance[i]}')
```

evaluasi fitur dari fitur-fitur yang terpilih

```python
def evaluate_model(x_train, x_test, y_train, y_test, model):
    model.fit(x_train, y_train)
    accuracy = model.score(x_test, y_test)
    return accuracy

```

```python
#model logistic regression untuk filter methods
logistic_model_filter = LogisticRegression(max_iter=200)
accuracy_filter = evaluate_model(x_train_filter, x_test_filter, y_train, y_test, logistic_model_filter)

#model logistic regression untuk wrapper method
logistic_model_rfe= LogisticRegression(max_iter=200)
accuracy_rfe = evaluate_model(x_train_rfe, x_test_rfe, y_train, y_test, logistic_model_rfe)

#model random forest untuk embeded method
accuracy_rf = evaluate_model(x_important, x_test_important, y_train, y_test, rf_model)
```

```python
print(f"Akurasi model dengan filter methods : {accuracy_filter:.2f}")
print(f"Akurasi model dengan wrapper method : { accuracy_rfe:.2f}")
print(f"Akurasi model random forest dengan embeded method {accuracy_rf}")
```

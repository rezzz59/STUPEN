```python
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Data: [Tinggi, Berat]
data_mentah = [
    [155, 50], # Andi
    [158, 52], # Budi
    [160, 55], # Caca
    [175, 80], # Dedi
    [178, 85], # Eko
    [180, 82]  # Feri
]

# Ubah jadi Tabel Rapi (DataFrame)
df = pd.DataFrame(data_mentah, columns=['Tinggi', 'Berat'])

# Tampilkan isinya
df
```

```python
df.info()
```

```python
df.describe()
```

```python
kmeans = KMeans(n_clusters = 2, random_state = 42)

```

```python
kmeans.fit(df)
```

```python
hasil = kmeans.labels_
print(hasil)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Tinggi'], df['Berat'], c=hasil, s=100, cmap='viridis')

plt.title("Kelompok kaos")
plt.ylabel('Tinggi Badan')
plt.xlabel('Berat Badan')
plt.show()
```

```python
centroid = kmeans.cluster_centers_
centroid
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(df['Tinggi'], df['Berat'], c = hasil)

#centroid
plt.scatter(centroid[:, 0], centroid[:, 1], c='red', s=200, marker= '+', label='Centroid')
plt.title("Peta ukuran kaos dan centroid")
plt.ylabel("Berat Badan")
plt.xlabel("Tinggi Badan")
plt.legend()
plt.grid()
plt.show()
```

```python
# Ambil data koordinat pusat
pusat = kmeans.cluster_centers_

print("=== SURAT PERINTAH PRODUKSI KAOS POLOS JAYA ===")
print(f"1. Ukuran KECIL (S) -> Jahit dengan Tinggi: {pusat[0][0]:.1f} cm, Berat: {pusat[0][1]:.1f} kg")
print(f"2. Ukuran BESAR (L) -> Jahit dengan Tinggi: {pusat[1][0]:.1f} cm, Berat: {pusat[1][1]:.1f} kg")
```
